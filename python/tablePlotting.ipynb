```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib notebook
```

```python
df = pd.read_csv('../24aa6454-044e-4a2a-9b29-c01b87f40da3policies.csv', header=None)
data = df.to_numpy()
data = np.split(data, len(data)/4400)

diff = []

for i in range(0, len(data) - 1):
    total = 0
    for j in range(0, len(data[0])):
        for k in range(0, len(data[0][0])):
            difference = abs(data[i][j][k] - data[i+1][j][k])
            total += difference
                
    diff.append(total)

print(data[0])
print(data[1])
plt.plot(diff)
```
